# Preprocessing our data

```python
def preprocess_data(new_data):
    new_data['Age'].fillna(new_data['Age'].mean(),inplace=True) # Completing Age column
    new_data.replace({'Sex':{'male': 1,'female':0}},inplace=True) # Convert the categories of the variable Sex
    new_data['Cabin']=new_data.Cabin.fillna('G6') #Completing Cabin column
    new_data.replace({'Survived':{'Yes': 1,'No':0}},inplace=True) # Convert the categories of the variable Survived
    new_data['Embarked'].fillna('S',inplace=True) # Completing Embarked column
    new_data['FamilySize']=new_data['SibSp']+new_data['Parch']
    return new_data
data=preprocess_data(df)
data
```

# Decision Tree Prediction

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score

#features extraction
x=data[["Pclass","Sex","Age","Fare","FamilySize"]]
y= data["Survived"]

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)  #splitting data with test size of 20%

#applying tree algorithm
d_tree = tree.DecisionTreeClassifier()  
d_tree.fit(x_train, y_train)  #fitting our model
y_pred=d_tree.predict(x_test)  # evaluating our model
print("accuracy:{}".format(accuracy_score(y_test, y_pred))) #print our accuracy
#We  got a classification rate of 76.53%, considered as good accuracy
```

# After Changing the decision tree parameters
